```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = DataFrame({'Sample':[0,1,2,3,4,5],
               'Animal': ['Dog','Dog','Cat','Cat','Dog', 'Cat'],
               'Intelligence': ['Dumb', 'Dumb', 'Smart', 'Smart', 'Smart', 'Dumb']})
df
```

```python
pd.crosstab(df.Animal, df.Intelligence, margins=True)
```
