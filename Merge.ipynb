```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = DataFrame({'key':['X','Z','Y','Z','X','X'], 'data_set_1':np.arange(6)})
df
```

```python
df2 = DataFrame({'key':['Q','Y','Z'], 'data_set_2':[1,2,3]})
df2
```

```python
pd.merge(df, df2)
```

```python
pd.merge(df,df2,on='key')
```

```python
pd.merge(df,df2,on='key',how='left')
```

```python
df2
```

```python
df
```

```python
pd.merge(df,df2,on='key',how='right')
```

```python
pd.merge(df,df2,on='key',how='outer')
```

```python
df_left = DataFrame({'key1':['SF','SF','LA'], 'key2':['one','two','three'], 'left_data':[10,20,30]})

df_left
```

```python
df_right = DataFrame({'key1':['SF','SF','LA','LA'], 'key2':['one','one','one', 'two'], 'right_data':[40,50,60, 70]})

df_right
```

```python
pd.merge(df_left, df_right,on=['key1','key2'], how='outer')
```
